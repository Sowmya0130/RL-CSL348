# Positive Corelation 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.randint(1, 101, size=100)
y = x + 100 + np.random.normal(0, 5, size=100)

df1 = pd.DataFrame({'x': x})
df2 = pd.DataFrame({'y': y})

df = pd.concat([df1, df2], axis=1)

df.plot.scatter(x='x', y='y')
plt.title('Scatter Plot of x and y')
plt.show()

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

# Negative Corelation 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(1, 101, size=100)
y = 100 - x + np.random.normal(0, 5, size=100)

df1 = pd.DataFrame({'x': x})
df2 = pd.DataFrame({'y': y})

df = pd.concat([df1, df2], axis=1)

df.plot.scatter(x='x', y='y')
plt.title('Scatter Plot of x and y')
plt.show()

corr_coef = np.corrcoef(df1['x'], df2['y'])[0, 1]
print("Pearson correlation coefficient:", corr_coef)

plt.scatter(df1['x'], df2['y'])
plt.title(f"Pearson correlation coefficient: {corr_coef:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import random 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x2 = []
for i in range(100):
    n = random.randint(1, 100)
    x2.append(n)

y2 = [107 - num for num in x2]

corr_coef = np.corrcoef(x2, y2)[0, 1]
print("Pearson correlation coefficient:", corr_coef)

plt.scatter(x2, y2)
plt.title(f"Pearson correlation coefficient: {corr_coef:.2f}")
plt.xlabel("x2")
plt.ylabel("y2")
plt.show()

df = pd.DataFrame({'x2': x2, 'y2': y2})

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import random 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x3 = []
for i in range(100):
    n = random.randint(1, 100)
    x3.append(n)

y3 = []
for i in range(100):
    n = random.randint(1, 100)
    y3.append(n)

corr_coef = np.corrcoef(x3, y3)[0, 1]
print("Pearson correlation coefficient:", corr_coef)

plt.scatter(x3, y3)
plt.title(f"Pearson correlation coefficient: {corr_coef:.2f}")
plt.xlabel("x3")
plt.ylabel("y3")
plt.show()

# Create a DataFrame from x3 and y3
df = pd.DataFrame({'x3': x3, 'y3': y3})

# Plot the colormap of the correlation matrix
sns.color_palette("coolwarm", as_cmap=True)
sns.heatmap(df.corr(), annot=True, cmap='magma')
plt.title('Correlation Colormap')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

random_array = np.random.randint(1, 200, size=100)
df1 = pd.DataFrame({'X': random_array})

random_array = np.random.randint(1, 200, size=100)
df2 = pd.DataFrame({'Y': random_array})

df = pd.concat([df1, df2], axis=1)

df.plot.scatter(x='X', y='Y')
plt.title('Scatter Plot of X and Y')
plt.show()

corr_matrix = df.corr()

plt.matshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.title('Correlation Matrix')
plt.show()
```

# Negative Corelation 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate x and y values with zero correlation
x = np.random.normal(0, 10, size=100)
y = np.random.normal(50, 10, size=100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Create a scatter plot
plt.scatter(df['x'], df['y'])
plt.title('Scatter Plot of x and y')
plt.show()

# Calculate the Pearson correlation coefficient
corr_coef = np.corrcoef(df['x'], df['y'])[0, 1]
print("Pearson correlation coefficient:", corr_coef)

# Display the correlation coefficient and scatter plot
plt.scatter(df['x'], df['y'])
plt.title(f"Pearson correlation coefficient: {corr_coef:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
